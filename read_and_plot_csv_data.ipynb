```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import csv
import pandas as pd
```

```python
data_path = "dataset/marks.csv"
```

```python
marks_csv = pd.read_csv(data_path)
```

```python
marks_csv
```

```python
subject = marks_csv.Subject
subject
```

```python
marks = marks_csv.Marks
marks
```

```python
plt.plot(subject,marks,color='maroon')
```

```python
xtitle = "Subjects"
ytitle = "Marks Obtaained"
```

```python
labels = subject
sizes = marks
```

```python
#Plotting the marks in pie charts

colors = ['lightgrey', 'yellow', 'lightblue', 'red', 'green', 'lightpink'] #Since we have 6 subjects

explode = (0.1, 0, 0, 0, 0, 0) #Specifying the exploding position of pie element

plt.pie(sizes,labels=labels,explode=explode,autopct='%1.2f%%', shadow=True, colors=colors)

plt.axis('equal')

plt.legend(title='Index', loc='upper right', bbox_to_anchor=(1,0,0.3,1))

plt.title('Marks Obtained in Pie Chart')

plt.show()
```

```python
# Now let us plot the data in bar-graph

fig = plt.figure(figsize=(10,5)) #height and width setup

plt.bar(labels,sizes,width=0.5,color=colors)

plt.xlabel(xtitle)

plt.ylabel(ytitle)

plt.title("Marks Obtained in Bar-Diagram")

plt.show()
```
